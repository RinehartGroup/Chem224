## Terminology

1. **Normal Modes**: A set of vibrational modes that are orthogonal to each other and are independent of each other.  The number of normal modes is equal to the number of degrees of freedom of the molecule.
2. **Degenerate Representation**: Irreducible representations that transform in the same way under all symmetry operations of the molecule.
      - Same character table entry and have the same eigenvalues.
      - Basis functions will form a degenerate subspace. (i.e. they are linearly dependent)
      - Degenerate representations are not necessarily the same as degenerate orbitals.
3. **Vibration**: A molecular motion that does not result in a net translation or rotation of the molecule.
4. **Internal Coordinates**: A set of coordinates that describe the position of atoms in a molecule relative to each other.  The number of internal coordinates is equal to the number of degrees of freedom of the molecule.
5. **Rovibrational Coupling**: The coupling of rotational and vibrational motions with the same symmetry in a molecule.
6. **Allowed Vibrational Transitions**: Transitions where the vibrational modes of the ground and excited states can form an eigenstate upon interaction with the electromagnetic field.
   1.  For IR spectroscopy, the transition dipole moment must be nonzero.
   2.  For Raman spectroscopy, the polarizability must be nonzero.
7.  **SALC**: Symmetry Adapted Linear Combination.  A linear combination of atomic orbitals that transforms according to an irreducible representation of the molecule.
8.  **Partner Function**: A function that transforms according to the same irreducible representation as a given function.  The partner function is orthogonal to the given function allowing for the construction of orthogonal linear combinations of basis fucntions that transforms according to the irreducible representation.
9.  **Projection operator**: A matrix that projects a vector onto a subspace.  The projection operator is the identity matrix for the subspace.
10. **Degrees of Freedom**: The number of independent ways a molecule can move.  The number of degrees of freedom is equal to the number of normal modes and the number of internal coordinates.
    1.  It is equal to the number of atoms times the number of dimensions minus the number of constraints.
        1.  In 3D, a molecule with $N$ atoms has $3N$ degrees of freedom. $3N - 6$ degrees of freedom are vibrational.
        2.  Linear molecules have 3N-5 degrees of freedom because their rotation is not quantized about the infinite rotation axis of the molecule (i.e. they can rotate about the axis without changing the energy of the molecule). 

## Key Equations

1. **Operator**: A function that maps a vector to another vector. 
   $$
   \hat{O} \psi = \psi'
   $$
   - $\hat{O}$ is the operator
   - $\psi$ is the input vector
   - $\psi'$ is the output vector

2. **Eigenfunction**: A function that is unchanged by an operator except for a scalar multiple.
      $$
      \hat{O} \psi = \lambda \psi
      $$
      - $\hat{O}$ is the operator
      - $\psi$ is the eigenfunction
      - $\lambda$ is the eigenvalue
   2. The eigenvalues of a representation are the characters of the representation.
   3. The eigenvalues of the Hamiltonian operator are the energies of the system.
3. **Projection Operator**: A matrix that projects a vector onto a subspace.
      $$
      P = \sum_{i=1}^{n} \psi_i \psi_i^*
      $$
      - $P$ is the projection operator
      - $\psi_i$ is the basis function
      - $n$ is the number of basis functions

4. **Character Projection Operator** (Projection Operator for a Representation): A matrix that projects a vector onto the subspace spanned by the basis functions of a representation. The projection operator can be written in terms of the characters of the representation:
      $$
      P = \frac{1}{h} \sum_{i=1}^{h} \chi_i \chi_i^{*}
      $$
   - $P$ is the projection operator
   - $\chi_i$ is the character of the representation
   - $h$ is the order of the group
   - $n$ is the number of basis functions
   - $i$ is the index of the basis function
   - $*$ indicates the complex conjugate
  
5. **Reduction Formula**: Used to reduce a reducible representation to a set of irreducible representations by projecting out the irreducible representations from the reducible representation.
      $$
      \Gamma = \sum_{i=1}^{h} n_i \chi_i
      $$

      - $\Gamma$ is the reducible representation
      - $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
      - $h$ is the order of the group (total number of operations in the group)
      - $i$ is the index of the irreducible representation
      - $\chi_i$ is the character of the irreducible representation
      - $*$ indicates the complex conjugate
 - To use the reduction formula, we usually know $\Gamma$ and $\chi_i$ and we want to solve for $n_i$. It can be rearranged to form:
  $$
   n_i = \frac{1}{h} \sum_{j=1}^{h} \Gamma_j \chi_i^{*}
   $$
   - $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
   - $h$ is the order of the group (total number of operations in the group)
   - $j$ is the index of the operation
   - $\Gamma_j$ is the character of the reducible representation for the $j^{th}$ operation
   - $\chi_i$ is the character of the irreducible representation
   - $*$ indicates the complex conjugate
 - The reduction formula is an example of using a projection operator where the basis functions are the irreducible representations of the group. We can describe the full form of the projection operator using the ket-bra:
$$
P = \sum_{i=1}^{h} \chi_i \chi_i^{*} = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right|
$$

- Then, we can represent our reducible representation as a linear combination of kets that are the irreducible representations of the group:
$$
\Gamma = \sum_{i=1}^{h} n_i \chi_i = \sum_{i=1}^{h} n_i \left| \chi_i \right>
$$

- Applying the projection operator to the reducible representation gives:
$$
P \Gamma = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right| \sum_{j=1}^{h} n_j \left| \chi_j \right> = \sum_{i=1}^{h} \sum_{j=1}^{h} n_j \left| \chi_i \right> \left< \chi_i | \chi_j \right> = \sum_{i=1}^{h} \sum_{j=1}^{h} n_j \left| \chi_i \right> \delta_{ij} = \sum_{i=1}^{h} n_i \left| \chi_i \right> = \Gamma
$$

- we are interested in where an atom is moving, not where it is
- therefore we can transform the displacement vector of the molecule and not the position vector 
- displacement vectors give us a very useful approximation:

$$
\vec{q_{k,x}} = \sqrt{m_k}(x_k - x_{k,0})
$$

  - $\vec{q_{k,x}}$ is the displacement vector for the $k^{th}$ atom in the $x$ dimension
  - $m_k$ is the mass of the $k^{th}$ atom
  - $x_k$ is the $x$ coordinate of the $k^{th}$ atom
  - $x_{k,0}$ is the equilibrium $x$ coordinate of the $k^{th}$ atom


- displacement vector removes dependence on the equilibrium position of the atom 
- only magnitude and direction matter - it becomes a vector!
- allows approximation of a harmonic potential for vibrations:

$$
V = \frac{1}{2} k \vec{q}^2
$$

- $V$ is the potential energy
- $\vec{q}$ is the displacement vector
- $k$ is the force constant which in the case of a vibration will simply be::

$$
k = \frac{\partial^2 V}{\partial \vec{q}^2} = m \omega^2
$$

- $m$ is the mass of the atom
- $\omega$ is the angular frequency of the vibration


The displacement vector for all molecular motions in the Cartesian basis can be written as:

$$
\vec{q} = \sum_{i=1,k=\{x,y,z\}}^{i=\frac{3N-6}{3}} \vec{q}_{k,i}
$$

- $\vec{q}$ is the displacement vector
- $N$ is the number of atoms in the molecule
- $i$ is the index of the atom
- $k$ is the index of the dimension
- $\vec{q}_{k,i}$ is the displacement vector for the $i^{th}$ atom in the $k^{th}$ dimension
- $x$, $y$, and $z$ are the Cartesian dimensions
- $3N-6$ is the number of vibrational degrees of freedom in a molecule with $N$ atoms

This will give a 3N dimensional column vector. For a molecule with three atoms for example, this will be a 9 dimensional column vector. For convenience, in written form, we can write this as the transpose of its conjugate (a 9 dimensional row vector):

$$
\vec{q} = \begin{bmatrix} q_{1,x} & q_{1,y} & q_{1,z} & q_{2,x} & q_{2,y} & q_{2,z} & q_{3,x} & q_{3,y} & q_{3,z} \end{bmatrix}^T
$$

Note that this displacement vector is *not* invariant under the symmetry of the molecule.

If we wanted to find how the vibrational modes interact directly from this displacement vector, we would need to solve for the overlap integral of the displacement vector with itself under the symmetry of the molecule.  This would be a 9x9 matrix.  Instead, we can use the projection operator to project out the irreducible representations of the group from the displacement vector. 

In the symmetry of a group, $G$, the displacement vector transforms according to the reducible representation, $\Gamma$:

$$
\Gamma = \sum_{i=1}^{h} n_i \chi_i
$$

- $\Gamma$ is the reducible representation
- $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
- $h$ is the order of the group (total number of operations in the group)
- $i$ is the index of the irreducible representation
- $\chi_i$ is the character of the irreducible representation
- $*$ indicates the complex conjugate

We can use the projection operator to project out the irreducible representations from the displacement vector:

$$
\Gamma = P \vec{q} = \sum_{i=1}^{h} \chi_i \chi_i^{*} \vec{q}
$$

- $\Gamma$ is the reducible representation
- $P$ is the projection operator
- $\vec{q}$ is the displacement vector
- $h$ is the order of the group (total number of operations in the group)
- $i$ is the index of the irreducible representation
- $\chi_i$ is the character of the irreducible representation
- $*$ indicates the complex conjugate
- $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$ c

The full 9x9 matrix form of the projection operator including all terms would be:

$$
P = \frac{1}{h} \sum_{i=1}^{h} \chi_i \chi_i^{*} = \frac{1}{h} \begin{bmatrix} \left| \chi_1 \right> \left< \chi_1 \right| & \left| \chi_1 \right> \left< \chi_2 \right| & \cdots & \left| \chi_1 \right> \left< \chi_h \right| \\ \left| \chi_2 \right> \left< \chi_1 \right| & \left| \chi_2 \right> \left< \chi_2 \right| & \cdots & \left| \chi_2 \right> \left< \chi_h \right| \\ \vdots & \vdots & \ddots & \vdots \\ \left| \chi_h \right> \left< \chi_1 \right| & \left| \chi_h \right> \left< \chi_2 \right| & \cdots & \left| \chi_h \right> \left< \chi_h \right| \end{bmatrix}
$$

but we know that only characters of the same irreducible representation are nonzero and so we can simplify this to:

$$
P = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right|
$$

- $P$ is the projection operator
- $\left| \chi_i \right>$ is the ket of the irreducible representation under symmetry operation i of the molecule
- $\left< \chi_i \right|$ is the bra of the irreducible representation under symmetry operation i of the molecule
- $h$ is the order of the group (total number of operations in the group)
- $i$ is the index of the irreducible representation

Notice that we are forming ket-bras and not bra-kets. This means that the projection operator will give us a matrix that is block diagonal in the irreducible representations of the group:

$$
P = \begin{bmatrix} \left| \chi_1 \right> \left< \chi_1 \right| & 0 & \cdots & 0 \\ 0 & \left| \chi_2 \right> \left< \chi_2 \right| & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & \left| \chi_h \right> \left< \chi_h \right| \end{bmatrix}
$$

- $P$ is the projection operator
- $\left| \chi_i \right>$ is the ket of the irreducible representation under symmetry operation i of the molecule
- $\left< \chi_i \right|$ is the bra of the irreducible representation under symmetry operation i of the molecule
- $h$ is the order of the group (total number of operations in the group)
- $i$ is the index of the irreducible representation
- $0$ is a zero matrix (this is necessary to generalize to multidimensional representations)
- $\cdots$ indicates that the matrix is block diagonal
- $*$ indicates the complex conjugate

Now, returning to our non-orthogonal displacement vector, we can project out the irreducible representations of the group using the projection operator:

$$
\Gamma = P \vec{q} = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right| \vec{q}
$$

or in matrix form:

$$
\Gamma = \begin{bmatrix} \left| \chi_1 \right> \left< \chi_1 \right| & 0 & \cdots & 0 \\ 0 & \left| \chi_2 \right> \left< \chi_2 \right| & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & \left| \chi_h \right> \left< \chi_h \right| \end{bmatrix} \begin{bmatrix} q_{1,x} \\ q_{1,y} \\ q_{1,z} \\ q_{2,x} \\ q_{2,y} \\ q_{2,z} \\ q_{3,x} \\ q_{3,y} \\ q_{3,z} \end{bmatrix}

## Example: Determine the symmetry of molecular motions of $H_{2}O$

### Using all molecular motions (translational, rotational, and vibrational)

#### Determine the total degrees of freedom of the molecule:

   1. $H_{2}O$ has 3 atoms and is non-linear:
      1. Degrees of freedom (total): $3 \times 3 = 9$
      2. Degrees of freedom (translational): $3$
      3. Degrees of freedom (rotational): $3$
      4. Degrees of freedom (vibrational): $9 - 3 - 3 = 3$
   
#### Determine the symmetry transformation characteristics that the molecule will conform to:

   1. The point group is $C_{2v}$.
   2. The character table is:
$$
\begin{array}{c|cccc}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) \\
\hline
A_{1} & 1 & 1 & 1 & 1 \\
A_{2} & 1 & 1 & -1 & -1 \\
B_{1} & 1 & -1 & 1 & -1 \\
B_{2} & 1 & -1 & -1 & 1 \\
\end{array}
$$

#### Find a reducible representation that describes all of the molecular motions:

 Since we will be using the characters of the irreducible representations, we only need to define the diagonal of the reducible representation. This means we just need to look for the basis elements that transform into themselves under the symmetry operations of the group. (Everything else will be "off diagonal" and thus not contribute to the diagonal of the reducible representation.)
 
 We can represent the atomic degrees of freedom by a set of vectors centered at each atomic site, leading to a 9 dimensional basis set.
$$
\begin{array}{c|c}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) \\
\hline
\Gamma_{red} & 9 & -1 & 3 & 1 \\
\hline
A_{1} & 1 & 1 & 1 & 1 \\
A_{2} & 1 & 1 & -1 & -1 \\
B_{1} & 1 & -1 & 1 & -1 \\
B_{2} & 1 & -1 & -1 & 1 \\
\end{array}
$$

#### The reduction formula

$$
   n_i = \frac{1}{h} \sum_{j=1}^{h} \Gamma_j \chi_i^{*}
$$
   - where $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
   - $h$ is the order of the group (total number of operations in the group)
   - $j$ is the index of the operation
   - $\Gamma_j$ is the character of the reducible representation for the $j^{th}$ operation
   - $\chi_i$ is the character of the irreducible representation
   - $*$ indicates the complex conjugate

#### Project out the irreducible representations using the reduction formula:

$$
\begin{array}{c|cccc|c}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) & n_i \\
\hline
\Gamma_{red} & 9 & -1 & 3 & 1 & 9 & \sum_{i=1}^{h} n_i \chi_i \\
\hline
A_{1} & 1 & 1 & 1 & 1 & 1 & ((1*9*1) + (1*-1*1) + (1*3*1) + (1*1*1))/4 = 3a_{1} \\
A_{2} & 1 & 1 & -1 & -1 & 1 & ((1*9*1) + (1*-1*1) + (1*3*-1) + (1*1*-1))/4 = 1a_{2} \\
B_{1} & 1 & -1 & 1 & -1 & 1 & ((1*9*1) + (1*-1*-1) + (1*3*1) + (1*1*-1))/4 = 3b_{1} \\
B_{2} & 1 & -1 & -1 & 1 & 1 & ((1*9*1) + (1*-1*-1) + (1*3*-1) + (1*1*1))/4 = 2b_{2} \\
\end{array}
$$

 - The reducible representation can therefore be decomposed into a direct sum of irreducible representations:

$$
\Gamma_{red} = 3a_{1} \oplus 1a_{2} \oplus 3b_{1} \oplus 2b_{2}
$$

 - Note: The direct sum is not the same as the sum of the characters. The direct sum is a linear combination of the irreducible representations. We can use this construction because the group representations form an algebra (i.e. they are closed under addition and multiplication) where addition is defined as the direct sum and multiplication is defined as the tensor or direct product.

#### Compare symmetries to identify translations, rotations, and vibrations

We can either determine the symmetry of the various types of motion by determining how they transform within the group and assigning them irreducible representations or by locating their respective functions on an existing character table.

$$
\begin{array}{c|cccc|cc}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) & linear & quadratic \\
\hline
A_1 & 1 & 1 & 1 & 1 & z & - \\
A_2 & 1 & 1 & -1 & -1 & - & R_z \\
B_1 & 1 & -1 & 1 & -1 & x & R_y \\
B_2 & 1 & -1 & -1 & 1 & y & R_x \\
\end{array}
$$

From the character table we can identify the following motions:

$$
\begin{array}{c|c}
\text{Motion} & \text{Irreducible Representation} \\
\hline
\text{All motions} & 3a_1 \oplus a_2 \oplus 3b_1 \oplus 2b_2 \\
{-} \quad \text{Translations} & a_1(z) \oplus b_1(x) \oplus b_2(y)  \\
{-} \quad \text{Rotations} & a_2(R_z) \oplus b_1(R_y) \oplus b_2(R_x) \\

### Using internal coordinates to find the symmetry of vibrations

Internal coordinates describe the position of the atoms relative to each other. This is quite useful for describing molecular vibrations since the internal coordinates are easy to put in terms of the symmetry. 

#### Parameterize the molecule by its degrees of freedom

Internal coordinates are perfect for describing bonds and angles.  To fully define it, we can use the following internal coordinates for $H_2O$:
   1. $r_{OH}$: The distance between the oxygen and hydrogen atom (there will be two of these)
   2. $\theta$: The angle between the two hydrogen atoms and the oxygen atom (H-O-H)

This allow us to define any internal displacement in terms of how it affects the internal coordinates.

   1. $\Delta \vec{r}_{OH(1)}$: The displacement vector between the oxygen and the first hydrogen atom
   2. $\Delta \vec{r}_{OH(2)}$: The displacement vector between the oxygen and the second hydrogen atom
   3. $\Delta \vec{\theta}_{H-O-H}$: The displacement vector between the two hydrogen atoms

#### Determine the symmetry of the internal coordinates (Displacement vectors)

If we focus on the bond displacement coordinates, $\Delta \vec{r}_{OH(1,2)}$, we can see that they likely form linear combinations representing the stretching motions of the molecule. Just like we did with the Cartesian basis, we can find a reducible representation of the bond displacement vectors and reduce it to its composition of irreducible representations.

$$
\begin{array}{c|cccc|c}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) & n_i \\
\hline
\Gamma_{red} & 2 & 0 & 2 & 0 & \sum_{i=1}^{h} n_i \chi_i \\
\hline
A_{1} & 1 & 1 & 1 & 1 & \frac{2 + 0 + 2 + 0}{4} = 1a_{1} \\
A_{2} & 1 & 1 & -1 & -1 & \frac{2 + 0 - 2 + 0}{4} = 0a_{2} \\
B_{1} & 1 & -1 & 1 & -1 & \frac{2 + 0 + 2 + 0}{4} = 1b_{1} \\
B_{2} & 1 & -1 & -1 & 1 & \frac{2 + 0 - 2 + 0}{4} = 0b_{2} \\
\end{array}
$$

 - The reducible representation of the stretching modes can therefore be decomposed into a direct sum of irreducible representations:

$$
\Gamma_{red} = 1a_{1} \oplus 1b_{1}
$$

   - We can do the same for the angle displacement coordinate, $\Delta \vec{\theta}_{H(1)-O-H(2)}$, to find its reducible representation:

$$
\begin{array}{c|cccc|c}
C_{2v} & E & C_{2} & \sigma_{v}(xz) & \sigma_{v}'(yz) & n_i \\
\hline
\Gamma_{red} & 1 & 1 & 1 & 1 & \sum_{i=1}^{h} n_i \chi_i \\
\hline
A_{1} & 1 & 1 & 1 & 1 & \frac{1 + 1 + 1 + 1}{4} = 1a_{1} \\
A_{2} & 1 & 1 & -1 & -1 & \frac{1 + 1 - 1 - 1}{4} = 0a_{2} \\
B_{1} & 1 & -1 & 1 & -1 & \frac{1 - 1 + 1 - 1}{4} = 0b_{1} \\
B_{2} & 1 & -1 & -1 & 1 & \frac{1 - 1 - 1 + 1}{4} = 0b_{2} \\
\end{array}
$$

   - The reducible representation of the angle bending mode can therefore be decomposed into a direct sum of irreducible representations:

$$
\Gamma_{red} = 1a_{1}
$$

#### Determine the form of the irreducible representations in terms of the internal coordinates

We have two irreducible representations for the stretching modes. How do we translate this into describing the stretching modes in terms of the internal coordinates? We can use the projection operator to project out the irreducible representations from the reducible representation.

The projection operator is the identity matrix for the subspace spanned by the irreducible representations. This means when we operate on our basis we will project it into that subspace. We can write the projection operator in terms of the characters of the irreducible representations:

$$
\hat{P} = \sum_{i=1}^{h} \chi_i \chi_i^{*} = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right|
$$

 - where $\hat{P}$ is the projection operator
 - $\chi_i$ is the character of the irreducible representation
 - $h$ is the order of the group (total number of operations in the group)
 - $i$ is the index of the irreducible representation
 - $*$ indicates the complex conjugate

In our case, we have two irreducible representations, $A_1$ and $B_1$ and they are both 1 dimensional. This means we can write the projection operator as:

$$
\hat{P} = \left| A_1 \right> \left< A_1 \right| \oplus \left| B_1 \right> \left< B_1 \right|
$$

The subspace we are projecting onto is the subspace spanning the irreducible representations. If there other components they will be "projected" out. This means we can separate the projecitons by symmetry so that $\hat{P}_{A_1}$ is the projection operator for the $A_1$ irreducible representation and $\hat{P}_{B_1}$ is the projection operator for the $B_1$ irreducible representation.

Our basis is the two stretching modes, $\Delta \vec{r}_{OH(1)}$ and $\Delta \vec{r}_{OH(2)}$. We can represent this as a ket vector composed of the basis vectors $\Delta \vec{r}_{OH(1)}$  and $\Delta \vec{r}_{OH(2)}$:

$$
\left| \Delta \vec{r}_{OH} \right> = \left| \Delta \vec{r}_{OH(1)} \right> \oplus \left| \Delta \vec{r}_{OH(2)} \right>
$$

and then apply the projection operator to it:

$$
\hat{P} \left| \Delta \vec{r}_{OH} \right> = \left| A_1 \right> \left< A_1 | \Delta \vec{r}_{OH} \right> + \left| B_1 \right> \left< B_1 | \Delta \vec{r}_{OH} \right> = \sum \left| \chi_i \right> \left< \chi_i | \Delta \vec{r}_{OH} \right>
$$

To show the projection operator in action, we can apply each symmetry operation on the basis and see how it transforms. Thus for $P_{A_1}$:

$$
P_{A_1} \left| \Delta \vec{r}_{OH} \right> = \hat{E} \left| \Delta \vec{r}_{OH} \right> \oplus \hat{C}_2 \left| \Delta \vec{r}_{OH} \right> \oplus \hat{\sigma}_{v}(xz) \left| \Delta \vec{r}_{OH} \right> \oplus \hat{\sigma}_{v}(yz) \left| \Delta \vec{r}_{OH} \right>
\\ \quad \\ 
= (1)\begin{bmatrix} 1 & 0 \\ 0 & 1 \\
\end{bmatrix} 
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(1)\begin{bmatrix} 0 & 1 \\ 1 & 0 \\
\end{bmatrix} 
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(1)\begin{bmatrix} 1 & 0 \\ 0 & 1 \\
\end{bmatrix} \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(1) \begin{bmatrix} 0 & 1 \\ 1 & 0 \\
\end{bmatrix} 
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} \\
\quad \\
= \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + \begin{bmatrix}
\Delta r_{OH(2)} \\
\Delta r_{OH(1)} \\
\end{bmatrix} + \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + \begin{bmatrix}
\Delta r_{OH(2)} \\
\Delta r_{OH(1)} \\
\end{bmatrix} \\
\quad \\
= 2 \begin{bmatrix}
\Delta r_{OH(1)} + \Delta r_{OH(2)} \\
\Delta r_{OH(2)} + \Delta r_{OH(1)} \\
\end{bmatrix} = 2 \left| \Delta \vec{r}_{OH(1)} + \Delta \vec{r}_{OH(2)} \right>
$$

Normalizing this mode gives:

$$
\nu_{A_1} = \frac{1}{\sqrt{2}} \left| \Delta \vec{r}_{OH(1)} + \Delta \vec{r}_{OH(2)} \right>
$$

   - This shows that the projection operator projects out the $A_1$ irreducible representation. Applied to the stretching basis vectors, it shows a stretching motion where the two bonds are moving in phase with each other. This is the symmetric stretching mode of the molecule.

For the $B_1$ Representation we get:

$$
P_{B_1} \left| \Delta \vec{r}_{OH} \right> = \hat{E} \left| \Delta \vec{r}_{OH} \right> \oplus \hat{C}_2 \left| \Delta \vec{r}_{OH} \right> \oplus \hat{\sigma}_{v}(xz) \left| \Delta \vec{r}_{OH} \right> \oplus \hat{\sigma}_{v}(yz) \left| \Delta \vec{r}_{OH} \right>
\\ \quad \\
= (1)\begin{bmatrix} 1 & 0 \\ 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(-1)\begin{bmatrix} 0 & 1 \\ 1 & 0 \\
\end{bmatrix}
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(1)\begin{bmatrix} 1 & 0 \\ 0 & 1 \\
\end{bmatrix} \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + 
(-1) \begin{bmatrix} 0 & 1 \\ 1 & 0 \\
\end{bmatrix}
\begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} \\
\quad \\
= \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + \begin{bmatrix}
-\Delta r_{OH(2)} \\
-\Delta r_{OH(1)} \\
\end{bmatrix} + \begin{bmatrix}
\Delta r_{OH(1)} \\
\Delta r_{OH(2)} \\
\end{bmatrix} + \begin{bmatrix}
-\Delta r_{OH(2)} \\
-\Delta r_{OH(1)} \\
\end{bmatrix} \\
\quad \\
= 2 \begin{bmatrix}
\Delta r_{OH(1)} - \Delta r_{OH(2)} \\
\Delta r_{OH(2)} - \Delta r_{OH(1)} \\
\end{bmatrix} = 2 \left| \Delta \vec{r}_{OH(1)} - \Delta \vec{r}_{OH(2)} \right>
$$

Normalizing this mode gives:

$$
\nu_{B_1} = \frac{1}{\sqrt{2}} \left| \Delta \vec{r}_{OH(1)} - \Delta \vec{r}_{OH(2)} \right>
$$

   - This shows that the projection operator projects out the $B_1$ irreducible representation. Applied to the stretching basis vectors, it shows a stretching motion where the two bonds are moving out of phase with each other. This is the anti-symmetric stretching mode of the molecule.
