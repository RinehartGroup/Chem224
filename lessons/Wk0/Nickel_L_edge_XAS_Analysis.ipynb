Figure 1. The wide energy range (a), and the expanded and non-resonant edge jump removed (b) Ni L-edge XAS spectra for complexes 1 (green) 2 (blue), and 3 (red). The L-edge XAS for a reference Ni(I) carbonyl compound (light grey) is shown for comparison (c). The figure shows features for L_3 and L_2. These edges can be broad and sometimes feature fine structures. In terms of Spectral Theory, the sharp peaks, such as those seen for the L_3 and L_2 features, correspond to the point spectrum and involve transitions to specific, localized unoccupied states. The overall 'edges' (such as the L-edge shown here) represent transitions to a continuum of unoccupied delocalized states, which can be interpreted through the continuous spectrum of spectral theory.
