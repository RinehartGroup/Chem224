### Terminology
   - *Point group* - A set of symmetry operations that can be performed that leave some object in an indistinguishable state. Point groups are groups that represent symmetries with respect to a point in space, where the space is a basis compatible with Cartesian space. *crystallogrpahic point group* a *point group* that is compatible with the lattice of a (usually 2D or 3D) crystal which further restricts the number of groups to  32 crystallographic point groups in 3D and 10 crystallographic point groups in 2D. 
   - *Symmetry operation* - A transformation that results in an indistinguishable state. 

Example: Say we have a homonuclear diatomic molecule, $H_2$. We can represent the position of the $H_2$ atoms however we want, but they should always be symmetric with respect to an axis splitting the bond between the two atoms. Let's say we name the hydrogen atoms $H_{(n)}$, $n=(1,2)$ and place them on a Cartesian coordinate system such that $H_{(1)}$ is at position $(\sqrt 2,\sqrt 2,0)$ and $H_{(2)}$ is at position $(-\sqrt 2,-\sqrt 2,0)$. This puts the molecular bond along the x-axis and the $C_2$ along the z-axis.
We can perform a symmetry operation on the molecule by rotating it about the $z$ axis by $\pi$ radians. This operation transforms the basis vectors as follows:

$$
|x \rangle \rightarrow -|x \rangle, \quad |y \rangle \rightarrow -|y \rangle, \quad and \quad |z \rangle \rightarrow |z \rangle
$$

and can thus be represented in matrix form: 

$$
\hat{R}(|x \rangle,|y \rangle,|z \rangle) \rarr (-|x \rangle,-|y \rangle,|z \rangle): \begin{bmatrix} x' \\ y' \\ z' \end{bmatrix} = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \\ z \end{bmatrix}
\hat{C}_{2_z} = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
$$

 Let's check if the molecule is in an indistinguishable state when we perform the operation $\hat{C}_2$ using the following definitions we just set up:

$$
\quad |H_{(1)} \rangle = \begin{bmatrix} \sqrt 2 \\ \sqrt 2 \\ 0 \end{bmatrix}, \quad  |H_{(2)} \rangle = \begin{bmatrix} -\sqrt 2 \\ -\sqrt 2 \\ 0 \end{bmatrix} , \quad \text{and} \quad \hat{R}(\pi) = \hat{C}_{2_z} = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
$$
and then apply the operator in matrix form to the basis set:
$$
\quad \hat{R}(\pi)|H_{(1)} \rangle = \hat{C}_{2_z}|H_{(1)} \rangle = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
\begin{bmatrix} \sqrt 2 \\ \sqrt 2 \\ 0 \end{bmatrix} = \begin{bmatrix} -\sqrt 2 \\ -\sqrt 2 \\ 0 \end{bmatrix} = H_{(2)}
$$
and for completeness, we can also check the other atom:
$$
\quad \hat{R}(\pi)|H_{(2)} \rangle = \hat{C}_{2_z}|H_{(2)} \rangle =
\begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
\begin{bmatrix} -\sqrt 2 \\ -\sqrt 2 \\ 0 \end{bmatrix} = \begin{bmatrix} \sqrt 2 \\ \sqrt 2 \\ 0 \end{bmatrix} = H_{(1)}
$$

this series of operations leaves the molecule in an indistinguishable state. In this case that indistinguishable state involves exchange of the two atoms:

$$
H_{(1)} \overset{\hat{C}_2}{\rarr} H_{(2)} 
$$

In this example I tried to make things more concrete by defining an external basis and adding a molecule to transform. This is important for looking at molecular symmetries, but if we just want to look at the symmetry of the group itself, that will be basis-independent. 

To check if $C_2$ is a symmetry operation of the group $C_2$ (stay with me here) we need to:
  - define its inverse
    - the inverse of a transformation is denoted by $\hat{A}^{-1}$
  - make sure its inverse is also a symmetry operation of the group. This means that \hat{A} and $\hat{A}^{-1}$ will be elements of the group. Note that often $\hat{A}^{-1} = \hat{A}$

In $C_2$, the inverse of $C_2$ is $C_2^{-1} = C_2$ and we can check that it is invariant under its inverse by applying the formula:
$$
\hat{A}^{-1}\hat{A} = \hat{A}\hat{A}^{-1} = \hat{I}
$$
where $\hat{I}$ is the identity operator. In this case, we can check that $C_2^{-1}C_2 = C_2C_2^{-1} = \hat{I}$ by applying the matrix representation of $C_2$ to itself:

$$
\hat{C}_2\hat{C}_2 = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix}\begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} = \hat{I}
$$

   - *Symmetry element* - A geometric object (line, plane, or point) about which a symmetry operation is performed. This becomes important when we need to define the group with respect to a particular basis set. For instance, if benzene is in the xy plane, then the $C_6$ axis is the z-axis. If benzene is in the xz plane, then the $C_6$ axis is the y-axis.
   - *Symmetry representation* - Any mathematical formalism that describes the transformation of an object under a symmetry operation. We can describe this mathematically by
$$
\hat{T} : \mathbb{R}^n \rightarrow \mathbb{R}^n, \quad \vec{v} \mapsto \vec{v'}
$$
where $\hat{T}$ is the transformation, $\mathbb{R}^n$ is a real vector space, $\vec{v}$ is a vector in the vector space, and $\vec{v'}$ is the transformed vector. In the context of quantum mechanics, the vector space is often expanded to a complex vector space, $\mathbb{C}^n$. 

Said aloud, this equation would read "the transformation, $\hat{T}$, takes a vector, $\vec{v}$, in the vector space, $\mathbb{R}^n$, and maps it to a new vector, $\vec{v'}$, in the vector space, $\mathbb{R}^n$."
   - *Matrix Representation* A mathematical formalism allowing us to algorithmically apply symmetry to a basis set of coordinates. Using matrices we can translate the abstract idea of a transformation into a computer-readible algorithm for applying the transformation to a set of coordinates. 
     
  If we find a matrix representation for the transformation \hat{T} we can then write the transformation as a matrix multiplication:

  $$
  \hat{T} : \mathbb{R}^n \rightarrow \mathbb{R}^n, \quad \vec{v} \mapsto \vec{v'} = \hat{T}\vec{v}
  $$

  where, for example, $\hat{T}$ could be a rotation matrix, $\hat{R}$, that rotates the vector, $\vec{v}$, by an angle, $\theta$, about the z-axis:

  $$
 \hat{R}(\theta)\vec{v} = \begin{bmatrix} \cos\theta & -\sin\theta & 0 \\ \sin\theta & \cos\theta & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \\ z \end{bmatrix} = \begin{bmatrix} x' \\ y' \\ z' \end{bmatrix} = \cos\theta\vec{x} + \sin\theta\vec{y}
  $$

   - *Basis set* - An inner product space (usually of vectors) that has properties suitable for representing wave functions. e.g. a set of atomic orbitals.
   - *Spectral theory* - general set of theorems that describe how to decompose a matrix into a set of eigenvectors and eigenvalues based on the symmetry of the matrix.
   - *Eigenvector* - A vector that is not changed by a transformation. The transformation only scales the vector by a constant. 
   - *Eigenvalue* - The constant by which an eigenvector is scaled under a transformation. In the context of quantum mechanics, a key eigenvalue is the energy of a system determined by the Hamiltonian operator, $\hat{H}$.
   - *Metric* - A function that defines the distance between two points in a space. In the context of quantum mechanics, $d(\psi, \phi) = \sqrt{\langle \psi - \phi | \psi - \phi \rangle}$, defines the metric on an inner product space describing the "distance" or dissimilarity between $\psi$ and $\phi$.
  
  In Cartesian space, the metric is the distance between two points also known as the magnitude of the vector connecting the two points or the norm:

  $$
  |\vec{r}| = \sqrt{x^2 + y^2 + z^2}
  $$

   - *Inner product* - A function that takes two vectors and returns a scalar. In the context of quantum mechanics, $\langle \psi | \phi \rangle$ represents taking the inner product of two wavefunctions. In Cartesian space, $\vec{r} \cdot \vec{r'} = x x' + y y' + z z'$ is the inner product of two vectors and represents a projection of one vector onto another.
   - *Outer product* - The outer product of two vectors $|a\rangle$ and $|b\rangle$ in a Hilbert space $\mathcal{H}$ is an operator $|a\rangle\langle b|$ that acts on the vectors in $\mathcal{H}$. Mathematically, it's defined using a ket-bra:

$$
|a\rangle\langle b| : \mathcal{H} \rightarrow \mathcal{H}
$$

where a and b are vectors in $\mathcal{H}$ such that the ket-bra is an invariant operator within the vector space of $\mathcal{H}$. In the context of quantum mechanics, the outer product of two vectors is often referred to as a *projection operator* as it projects the vector $|c\rangle$ onto the vector $|b\rangle$ and then scales the result by the magnitude of the vector $|a\rangle$. For instance, if

$$
|a\rangle = |b\rangle = |c\rangle = |x\rangle \quad then \quad |a\rangle\langle b|c\rangle = |x\rangle\langle x|x\rangle = |x\rangle\langle x|x\rangle = |x\rangle
$$

Example: Start with the basis set defined by: 

$$
({|p_x\rangle, |p_y\rangle, |p_z\rangle}) \quad \epsilon \quad \mathcal{V}
$$

 We can form a projection operator, $\hat{P}$, that projects a vector, $|c\rangle$, onto each vector $\langle p_i|$ and then scales the result by the magnitude of the vector : $|p_i \rangle$. We can represent the projection operator as:
$$
\hat{P} = \sum{|p_i \rangle \langle p_i|}
$$

and the action of the projection operator on a vector, $|c\rangle$, as:

$$ 
\hat{P}|c\rangle = \sum{|p_i \rangle \langle p_i|c\rangle}
$$

and, in the p-orbital basis set, the projection operator becomes:
$$
\sum{|p_i \rangle \langle p_j|} = \begin{bmatrix} |p_x \rangle \langle p_x| & |p_x \rangle \langle p_y| & |p_x \rangle \langle p_z| \\ |p_y \rangle \langle p_x| & |p_y \rangle \langle p_y| & |p_y \rangle \langle p_z| \\ |p_z \rangle \langle p_x| & |p_z \rangle \langle p_y| & |p_z \rangle \langle p_z| \end{bmatrix}
$$

If we define the vector $|c\rangle$ as 

$$
|c\rangle = \begin{bmatrix} |p_x \rangle \\ |p_y \rangle \\ |p_z \rangle \end{bmatrix}
$$

then the action of the projection operator on $|c\rangle$ becomes:
$$
\hat{P}|c\rangle = \begin{bmatrix} |p_x \rangle \langle p_x| & |p_x \rangle \langle p_y| & |p_x \rangle \langle p_z| \\ |p_y \rangle \langle p_x| & |p_y \rangle \langle p_y| & |p_y \rangle \langle p_z| \\ |p_z \rangle \langle p_x| & |p_z \rangle \langle p_y| & |p_z \rangle \langle p_z| \end{bmatrix} \begin{bmatrix} |p_x \rangle \\ |p_y \rangle \\ |p_z \rangle \end{bmatrix} = \begin{bmatrix} |p_x \rangle \langle p_x|p_x \rangle + |p_x \rangle \langle p_y|p_y \rangle + |p_x \rangle \langle p_z|p_z \rangle \\ |p_y \rangle \langle p_x|p_x \rangle + |p_y \rangle \langle p_y|p_y \rangle + |p_y \rangle \langle p_z|p_z \rangle \\ |p_z \rangle \langle p_x|p_x \rangle + |p_z \rangle \langle p_y|p_y \rangle + |p_z \rangle \langle p_z|p_z \rangle \end{bmatrix} = \begin{bmatrix} |p_x \rangle \\ |p_y \rangle \\ |p_z \rangle \end{bmatrix} = |c\rangle
$$

where now we can define the projection operator in terms of the basis set vectors since we now have a representation of $\hat{P}|c\rangle$ in terms of the basis set with scalar coefficients. This means that we can simplify the projection operator as a matrix:
$$
\hat{P} = \sum{|p_i \rangle \langle p_j|} = \begin{bmatrix} |p_x \rangle \langle p_x| & |p_x \rangle \langle p_y| & |p_x \rangle \langle p_z| \\ |p_y \rangle \langle p_x| & |p_y \rangle \langle p_y| & |p_y \rangle \langle p_z| \\ |p_z \rangle \langle p_x| & |p_z \rangle \langle p_y| & |p_z \rangle \langle p_z| \end{bmatrix} = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} = \mathbb{I}_3
$$
or simply say that:
$$
|p_i \rangle \langle p_j| = 1 \quad \text{if} \quad i = j \quad \text{and} \quad |p_i \rangle \langle p_j| = 0 \quad \text{if} \quad i \neq j
$$
This behavior is that of the Kronecker delta function, $\delta_{ij}$, which is defined as:
$$
\delta_{ij} = 1 \quad \text{if} \quad i = j \quad \text{and} \quad \delta_{ij} = 0 \quad \text{if} \quad i \neq j
$$

so this version of the *projection operator* can be written as:
$$
\hat{P} = \sum{|p_i \rangle \langle p_j|} = \sum{\delta_{ij}} = \mathbb{I}_3
$$

   - *Eigenspace* - The set of all eigenvectors of a transformation. In the context of quantum mechanics, the eigenspace of an operator, $\hat{A}$, is the set of all vectors, $|a\rangle$, such that $\hat{A}|a\rangle = a|a\rangle$ where $a$ is the eigenvalue of the eigenvector, $|a\rangle$.

   - *Eigenspace Projection* - A *projection operator* that projects a vector onto an eigenspace. In the context of quantum mechanics, the projection operator, $\hat{P}_a$, projects a vector, $|c\rangle$, onto the eigenspace of an operator, $\hat{A}$, with eigenvalue, $a$:
$$
 \mathcal{P} \vec{v} = \frac{\vec{v} \cdot \vec{u}}{\vec{u} \cdot \vec{u}} \vec{u} = \frac{\vec{v} \cdot \vec{u}}{|\vec{u}|^2} \vec{u} = \frac{\vec{v} \cdot \vec{u}}{|\vec{u}|} \hat{u}
$$

which indicates that the projection of $\vec{v}$ onto $\vec{u}$ is the scalar projection of $\vec{v}$ onto $\vec{u}$ times the unit vector $\hat{u}$. This is the same as the projection of $\vec{v}$ onto $\vec{u}$ times the unit vector $\hat{u}$. 

## Matrix Representations of Symmetry Operations

### General Symmetry Transformation
  
  In general, a symmetry operation is a transformation that can be described by a matrix. Here, and in many of the simple examples we will look at early on, we define our transformation using the Cartesian coordinate system. In this system, a point in space can be described by a set of three coordinates, $x,y,z$ where each coordinate is a real number,  mathspeak: $x, y, z \in \mathbb{R}$. Another way to describe this is to use a vector, $\vec{r}$, where $\vec{r} = (x,y,z)$ and $\vec{r} \in \mathbb{R}^3$. This description is very useful for looking at the mathematical formalism of symmetry operations. For instance, we can describe the Cartesian space using the unit vectors, $\hat{x}, \hat{y}, \hat{z}$, where $\hat{x} = (1,0,0)$, $\hat{y} = (0,1,0)$, and $\hat{z} = (0,0,1)$ which is a basis set for the Cartesian space because any point in the Cartesian space can be described as a linear combination of these unit vectors. For instance, $\vec{r} = x\hat{x} + y\hat{y} + z\hat{z}$. We can also describe the vector $\vec{r}$ as a magnitude, $|\vec{r}| = \sqrt{x^2 + y^2 + z^2}$ which defines the metric of the space meaning that there is a well defined meaning to the idea of distance in the space. We can also define a unit vector, $\hat{r}$, that points in the direction of $\vec{r}$, where $\hat{r} = \frac{\vec{r}}{|\vec{r}|}$. This means that any vector has a direction independent of its magnitude meaning it can be normalized and scaled independently (two vital properties of a basis set and an eigenspace).
    
  A symmetry operation is a transformation that takes a point in space and transforms it to a new point in space. The transformation is described by a matrix, $T$, that takes the coordinates of the original point and transforms them to the coordinates of the new point. The matrix is applied to the coordinates of the original point as follows:

  $$T:(x,y,z) \rightarrow (x',y',z')$$ where $x',y',z'$ are the coordinates of the transformed point. In matrix form, $$\begin{bmatrix} x' \\ y' \\ z' \end{bmatrix} = \begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix} \begin{bmatrix} x \\ y \\ z \end{bmatrix}$$

## Rotational Transformations

Unlike mirror symmetry, rotational transformations retain the "sense" of the object in the space. This means that they can be either clockwise (-) or counterclockwise (+). The rotation axis is a unit vector, $\hat{n} = (n_x, n_y, n_z)$, where $n_x^2 + n_y^2 + n_z^2 = 1$. Point Group symmetry uses rotation axes as the primary means to classify groups meaning the highest n rotation axis is the primary axis of symmetry whereby other symmetry elements are defined. Sometimes this axis will be $C_1$.

In the most general case for 3D cartesian coordinates, a rotation about an axis, $\hat{n}$, by an angle, $\theta$, is defined as:

  $$R(\theta) = \begin{bmatrix} \cos\theta & -\sin\theta & 0 \\ \sin\theta & \cos\theta & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

### Powers of Rotations

  The powers of a rotation are defined as follows:
  
  $$R^n(\theta) = \underbrace{R(\theta)R(\theta)R(\theta)...R(\theta)}_{n \text{ times}}$$
  
  For instance, $R^2(\theta) = R(\theta)R(\theta)$.
  
  The powers of a rotation are also rotations. This means that they can be represented by a matrix. For instance, $R^2(\theta)$ is a rotation by $2\theta$ and can be represented by the matrix:
  
  $$R^2(\theta) = \begin{bmatrix} \cos(2\theta) & -\sin(2\theta) & 0 \\ \sin(2\theta) & \cos(2\theta) & 0 \\ 0 & 0 & 1 \end{bmatrix}$$
  
  In general, the powers of a rotation can be represented by the matrix:
  
  $$R^n(\theta) = \begin{bmatrix} \cos(n\theta) & -\sin(n\theta) & 0 \\ \sin(n\theta) & \cos(n\theta) & 0 \\ 0 & 0 & 1 \end{bmatrix}$$
  
  The powers of R other than R itself are reducible and can always be represented by another, simpler matrix. See the table below for examples. 

| $C_n$ | $R^1$ | $R^2$ | $R^3$ | $R^4$ | $R^5$ | $R^6$ |
| --- | --- | --- | --- | --- | --- | ---|
| $C_2$ | $$\begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix}$$ | $$\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = E$$ | $$\begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix} = C_2^1$$ | $$\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = E$$ | $$\begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix} = C_2^1$$ | $$\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = E$$ |
| $C_6$ | $$\begin{bmatrix} 1/2 & -\sqrt{3}/2 \\ \sqrt{3}/2 & 1/2 \end{bmatrix} = C_6^1$$ | $$\begin{bmatrix} -1/2 & -\sqrt{3}/2 \\ \sqrt{3}/2 & -1/2 \end{bmatrix} = C_6^2 = C_3^1 $$ | $$\begin{bmatrix} -1 & 0 \\ 0 & -1 \end{bmatrix} = C_6^3 = C_2^1 $$ | $$\begin{bmatrix} -1/2 & \sqrt{3}/2 \\ -\sqrt{3}/2 & -1/2 \end{bmatrix} = C_6^4 = C_3^2 $$ | $$\begin{bmatrix} 1/2 & \sqrt{3}/2 \\ -\sqrt{3}/2 & 1/2 \end{bmatrix} = C_6^5 = C6^-1 $$ | $$\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = C_6^6 = E $$ |


## Inversion Transformation:

A center of inversion acts to invert all coordinates $(x,y,z)$ to their opposite direction $(-x,-y,-z)$. In quantum field theory this is refered to as a parity transformation and represents one of the fundamental symmetries of the universe (the "P" in CPT symmetry) as currently understood by the standard model.

| Position | Example | Transformation | Transformation Matrix |
| --- | --- | --- | --- |
| General | $i$ | $i:(x,y,z) \rightarrow (-x,-y,-z)$ | $$\begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & -1 \end{bmatrix}$$ |
| on Origin | $i$ | $i:(0,0,0) \rightarrow (0,0,0)$ | $$\begin{bmatrix} 0 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & 0 \end{bmatrix} = 0 $$ |

## Identity Transformation:

The identity transformation is the transformation that leaves all points in space unchanged. It is represented by the identity matrix, $\mathbb{I}_3$ or the letter $E$ in Schoenflies notation. Every symmetry group has an identity transformation and every symmetry element has an inverse element such that the product of the two is the identity transformation (just like how the product of a number and its reciprocal is 1).

### The Identity Element and Inverses: 

  - Every symmetry element has an inverse such that: $$T^{-1}T = TT^{-1} = E$$
  - The inverse of a rotation is the rotation in the opposite direction: $$R^{-1}(\theta) = R(-\theta)$$
  - The inverse of a mirror plane is the same mirror plane: $$\sigma^{-1} = \sigma$$
  - The inverse of the identity transformation is the identity transformation: $$E^{-1} = E$$
  - The inverse of the inversion transformation is the inversion transformation: $$i^{-1} = i$$
  - Rotoreflection inverses are easiest to define by simply equating them to the product of the inverse of the rotation and the mirror plane: $$S_n^{-1} = C_n^{-1}\sigma$$
    - Be careful with odd rotoreflections as the inverse element $S_n^{-1}$ is not the same as $S_n^{n-1} (e.g. S_3^2 \neq S_3^{-1})$.

## Rotoreflection Transformations (Improper Rotations):

Rotoreflections are a combination of a rotation and a reflection. They are represented by the matrix product of the rotation and the reflection. For instance, a $S_4$ rotation is a rotation by $\frac{2\pi}{4} = \frac{\pi}{2}$ followed by a reflection in the mirror plane perpendicular to the $C_4$ axis. 

### Matrix Representations of Rotoreflection Transformations:

The matrix representation of a rotoreflection is the product of the matrix representations of the rotation and the reflection. For instance, the matrix representation of $S_4$ is:

$$S_4 = R(\frac{\pi}{2})\sigma_{h} = \begin{bmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & 1 \end{bmatrix}\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & -1 \end{bmatrix} = \begin{bmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & -1 \end{bmatrix}$$

We can extend the matrix representation of a rotoreflection to any power of $S_n^x$ by using the matrix representation of the composite transformation and taking the matrix power just as we did for the rotation. For instance, the matrix representation of $S_4^2$ is:

$$S_4^2 = R(\frac{\pi}{2})\sigma_{h}R(\frac{\pi}{2})\sigma_{h} = \begin{bmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & -1 \end{bmatrix}\begin{bmatrix} 0 & -1 & 0 \\ 1 & 0 & 0 \\ 0 & 0 & -1 \end{bmatrix} = \begin{bmatrix} -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & 1 \end{bmatrix} = C_2$$
