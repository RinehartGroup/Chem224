## Multiplication Tables for Groups

### The Problem

How can we determine if some symmetry operations form a group?
 - Form the multiplication table for the group.
 - Check if the group axioms are satisfied.

### Example: $C_2$

Proposition: the set of two elements: $\{ e, C_2^1 \}$ forms a group.

To be a group, the *Group Axioms* must be satisfied. It helps to form the multiplication table for all the symmetry operations we think may be in the group:

| $\cdot$ | $e$ | $C_2^1$ |
| --- | --- | --- |
| $e$ | $e$ | $C_2^1$ |
| $C_2^1$ | $C_2^1$ | $e$ |

#### Group Axioms
    
1) *Closure*: 
    - definition: for all $g, h \in G$, $g \cdot h \in G$

Check in $C_2$:
- $e \cdot e = e \in C_2$
- $e \cdot C_2^1 = C_2^1 \in C_2$

2) *Associativity*: 
   - definition: for all $g, h, k \in G$, $(g \cdot h) \cdot k = g \cdot (h \cdot k)$

Check in $C_2$:
- $(e \cdot e) \cdot e = e \cdot e = e$
- $e \cdot (e \cdot e) = e \cdot e = e$
- $(e \cdot C_2^1) \cdot e = C_2^1 \cdot e = C_2^1$

3) *Identity*: 
   - definition: there exists $e \in G$ such that for all $g \in G$, $e \cdot g = g \cdot e = g$

Check in $C_2$:
- $e \cdot e = e$
- $C_2^1 \cdot e = C_2^1$
- $e \cdot C_2^1 = C_2^1$

1) *Inverse*: 
   - definition: for all $g \in G$, there exists $h \in G$ such that $g \cdot h = h \cdot g = e$
  
Check in $C_2$:
- $e \cdot e = e$
- $C_2^1 \cdot C_2^1 = e$

The symmetry operations $\{ e, C_2^1 \}$ form a group because they satisfy the group axioms.

## The Group Algebra

*Group Algebra*: the set of all linear combinations of the group elements with complex coefficients.

### The Algebra $\mathbb{C}[C_2]$

 - Group: set of elements with a binary operation
    - In this case, the group is $C_2 = \{ e, C_2^1 \}$
    - The binary operation is multiplication (also called composition). If we are using matrix representations, the binary operation is the dot (inner) product.
    - A group does not have other operations, such as scalar multiplication or addition.

Group Algebra: the set of all linear combinations of the group elements with complex coefficients.

 - In this case, the group algebra is $\mathbb{C}[C_2] = \{ a e + b C_2^1 \mid a, b \in \mathbb{C} \}$
 - The group algebra is a vector space, so it has operations such as scalar multiplication and addition.
 - We can use the group algebra to perform operations on representations of the group elements, but first we need to define the representations in a useful way.

## Decomposing a Group Algebra Using Projection Operators

### The Problem

- For the group $C_2$, the symmetry operations $\{ e, C_2^1 \}$ form a group but not a good basis for the group algebra $\mathbb{C}[C_2]$. We're gonna need a better basis.
- We can decompose the group algebra $\mathbb{C}[C_2]$ into a direct sum of subspaces. This is called *decomposition of the group algebra*.

### The Group Algebra $\mathbb{C}[C_2]$

Elements $x$ in $\mathbb{C}[C_2]$ can be expressed as linear combinations of the group elements:

$$
x = \frac{1}{2}(ax_e + bx_a)
$$

### Projection Operator

The projection operator $P(x)$ is defined as:

$$
P(x) = \frac{1}{|C_2|} \sum_{g \in C_2} g \cdot x \cdot g^{-1}
$$

For $C_2$, this simplifies to:

$$
P(x) = \frac{1}{2} \left( e \cdot \left(\frac{1}{2}(x_e + x_a)\right) \cdot e^{-1} + a \cdot \left(\frac{1}{2}(x_e + x_a)\right) \cdot a^{-1} \right)
$$

### Finding Idempotents

Upon applying the projection operator, we find:

$$
P(x) = \frac{1}{2} x
$$

This leads us to consider symmetric and antisymmetric combinations as idempotents:

1. Symmetric: $\frac{1}{2}(x_e + x_a)$
2. Antisymmetric: $\frac{1}{2}(x_e - x_a)$

### Conclusions

- The projection operator approach provides a way to decompose the group algebra without using character theory.
- Symmetric and antisymmetric combinations emerge as idempotents in this specific case.

## Relating to Dirac Notation

In Dirac notation, projection operators can be expressed as:

$$
P = | \psi \rangle \langle \psi |
$$

Idempotents in this context would correspond to states $| \psi \rangle$ such that $P^2 = P$, or $| \psi \rangle \langle \psi | \psi \rangle \langle \psi | = | \psi \rangle \langle \psi |$.

For $C_2$, the symmetric and antisymmetric combinations could be thought of as these $| \psi \rangle$ states, making the corresponding projection operators idempotent.

Certainly! Let's consider a simple function $f$ defined on the group $C_2$, which consists of two elements $\{ e, a \}$. In Dirac notation, we can represent $f$ as a ket $| f \rangle$.

### Function $f$ in Original Basis

To split a function $f$ into its components $f_e$ and $f_a$ corresponding to the elements $e$ and $a$ of the group $C_2$, you would evaluate $f$ at these group elements. Essentially, $f_e$ and $f_a$ are the values of the function $f$ at the identity element $e$ and the other element $a$, respectively.

### Mathematical Representation

Let's say $f: C_2 \rightarrow \mathbb{C}$ is the function defined on the group $C_2$. Then:

- $f_e = f(e)$
- $f_a = f(a)$

### Example

Suppose $f(g) = g^2$ for $g \in C_2$ (where $e = 1$ and $a = -1$).

- $f_e = f(1) = 1^2 = 1$
- $f_a = f(-1) = (-1)^2 = 1$

### In Matrix Form

Once you have $f_e$ and $f_a$, you can represent $f$ as a column vector:

$$
| f \rangle = \begin{pmatrix} f_e \\ f_a \end{pmatrix}
$$

In our example, this would be:

$$
| f \rangle = \begin{pmatrix} 1 \\ 1 \end{pmatrix}
$$

### Projection Operators

We've identified two idempotents (symmetric and antisymmetric combinations) that correspond to projection operators:

1. Symmetric: $P_{\text{sym}} = \frac{1}{2} (| e \rangle \langle e | + | a \rangle \langle a |)$
2. Antisymmetric: $P_{\text{anti}} = \frac{1}{2} (| e \rangle \langle e | - | a \rangle \langle a |)$

### Changing the Basis

To change the basis of $f$, we apply these projection operators:

1. Symmetric part of $f$:
$$
| f_{\text{sym}} \rangle = P_{\text{sym}} | f \rangle = \frac{1}{2} (f_e + f_a) | e \rangle + \frac{1}{2} (f_e + f_a) | a \rangle
$$

2. Antisymmetric part of $f$:
$$
| f_{\text{anti}} \rangle = P_{\text{anti}} | f \rangle = \frac{1}{2} (f_e - f_a) | e \rangle + \frac{1}{2} (f_e - f_a) | a \rangle
$$

### Function $f$ in New Basis

In the new basis, $f$ can be expressed as the sum of its symmetric and antisymmetric parts:

$$
| f \rangle = | f_{\text{sym}} \rangle + | f_{\text{anti}} \rangle
$$

### Function $f$ in Original Basis

Let's assume $f$ is represented as a column vector in the original basis:

$$
| f \rangle = \begin{pmatrix} f_e \\ f_a \end{pmatrix}
$$

### Projection Operators in Matrix Form

The projection operators in matrix form are:

1. Symmetric: $P_{\text{sym}} = \frac{1}{2} \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$
2. Antisymmetric: $P_{\text{anti}} = \frac{1}{2} \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$

### Changing the Basis

To change the basis of $f$, we apply these projection operators:

1. Symmetric part of $f$:

$$
| f_{\text{sym}} \rangle = P_{\text{sym}} | f \rangle = \frac{1}{2} \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} f_e \\ f_a \end{pmatrix} = \begin{pmatrix} \frac{1}{2} f_e \\ \frac{1}{2} f_a \end{pmatrix}
$$

2. Antisymmetric part of $f$:

$$
| f_{\text{anti}} \rangle = P_{\text{anti}} | f \rangle = \frac{1}{2} \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix} \begin{pmatrix} f_e \\ f_a \end{pmatrix} = \begin{pmatrix} \frac{1}{2} f_e \\ -\frac{1}{2} f_a \end{pmatrix}
$$

### Function $f$ in New Basis

In the new basis, $f$ can be expressed as the sum of its symmetric and antisymmetric parts:

$$
| f \rangle = | f_{\text{sym}} \rangle + | f_{\text{anti}} \rangle = \begin{pmatrix} \frac{1}{2} f_e \\ \frac{1}{2} f_a \end{pmatrix} + \begin{pmatrix} \frac{1}{2} f_e \\ -\frac{1}{2} f_a \end{pmatrix} = \begin{pmatrix} f_e \\ 0 \end{pmatrix}
$$

This shows how the projection operators help us decompose $f$ into its symmetric and antisymmetric components in the matrix representation.
