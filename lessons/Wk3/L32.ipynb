## Key Equations

1. **Operator**: A function that maps a vector to another vector. 
   $$
   \hat{O} \psi = \psi'
   $$
   - $\hat{O}$ is the operator
   - $\psi$ is the input vector
   - $\psi'$ is the output vector

2. **Eigenfunction**: A function that is unchanged by an operator except for a scalar multiple.
      $$
      \hat{O} \psi = \lambda \psi
      $$
      - $\hat{O}$ is the operator
      - $\psi$ is the eigenfunction
      - $\lambda$ is the eigenvalue
   2. The eigenvalues of a representation are the characters of the representation.
   3. The eigenvalues of the Hamiltonian operator are the energies of the system.
3. **Projection Operator**: A matrix that projects a vector onto a subspace.
      $$
      P = \sum_{i=1}^{n} \psi_i \psi_i^*
      $$
      - $P$ is the projection operator
      - $\psi_i$ is the basis function
      - $n$ is the number of basis functions

4. **Character Projection Operator** (Projection Operator for a Representation): A matrix that projects a vector onto the subspace spanned by the basis functions of a representation. The projection operator can be written in terms of the characters of the representation:
      $$
      P = \frac{1}{h} \sum_{i=1}^{h} \chi_i \chi_i^{*}
      $$
   - $P$ is the projection operator
   - $\chi_i$ is the character of the representation
   - $h$ is the order of the group
   - $n$ is the number of basis functions
   - $i$ is the index of the basis function
   - $*$ indicates the complex conjugate
  
5. **Reduction Formula**: Used to reduce a reducible representation to a set of irreducible representations by projecting out the irreducible representations from the reducible representation.
      $$
      \Gamma = \sum_{i=1}^{h} n_i \chi_i
      $$

      - $\Gamma$ is the reducible representation
      - $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
      - $h$ is the order of the group (total number of operations in the group)
      - $i$ is the index of the irreducible representation
      - $\chi_i$ is the character of the irreducible representation
      - $*$ indicates the complex conjugate
 - To use the reduction formula, we usually know $\Gamma$ and $\chi_i$ and we want to solve for $n_i$. It can be rearranged to form:
  $$
   n_i = \frac{1}{h} \sum_{j=1}^{h} \Gamma_j \chi_i^{*}
   $$
   - $n_i$ is the number of times the irreducible representation $\chi_i$ appears in the reducible representation, $\Gamma$
   - $h$ is the order of the group (total number of operations in the group)
   - $j$ is the index of the operation
   - $\Gamma_j$ is the character of the reducible representation for the $j^{th}$ operation
   - $\chi_i$ is the character of the irreducible representation
   - $*$ indicates the complex conjugate
 - The reduction formula is an example of using a projection operator where the basis functions are the irreducible representations of the group. We can describe the full form of the projection operator using the ket-bra:
$$
P = \sum_{i=1}^{h} \chi_i \chi_i^{*} = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right|
$$

- Then, we can represent our reducible representation as a linear combination of kets that are the irreducible representations of the group:
$$
\Gamma = \sum_{i=1}^{h} n_i \chi_i = \sum_{i=1}^{h} n_i \left| \chi_i \right>
$$

- Applying the projection operator to the reducible representation gives:
$$
P \Gamma = \sum_{i=1}^{h} \left| \chi_i \right> \left< \chi_i \right| \sum_{j=1}^{h} n_j \left| \chi_j \right> = \sum_{i=1}^{h} \sum_{j=1}^{h} n_j \left| \chi_i \right> \left< \chi_i | \chi_j \right> = \sum_{i=1}^{h} \sum_{j=1}^{h} n_j \left| \chi_i \right> \delta_{ij} = \sum_{i=1}^{h} n_i \left| \chi_i \right> = \Gamma
$$


  
## Example: Selection Rules for Vibrational Modes (IR Spectroscopy)

### The Dipole Moment Operator

The dipole moment operator is a vector operator that describes the electric dipole moment of a molecule. It is defined as:

$$
\hat{\mu} = \sum_{i=1}^{N} q_i \vec{r}_i
$$

 - where $\hat{\mu}$ is the dipole moment operator
 - $N$ is the number of atoms in the molecule
 - $q_i$ is the charge of the $i^{th}$ atom
 - $\vec{r}_i$ is the position vector of the $i^{th}$ atom

The dipole moment operator is used to describe the "allowedness" of vibrational transitions under IR light. It represents the operator that couples the vibrational states to the electromagnetic field. The operator acts on vibrational states due to the fact that vibrational motions redistribute the charge density of the molecule. 

$$
\hat{\mu} \psi = \psi'
$$

 - where $\hat{\mu}$ is the dipole moment operator
 - $\psi$ is the vibrational state
 - $\psi'$ is the new vibrational state

We know that vibrational states $\psi_i$ and $\psi_j$ can only overlap if they have the same symmetry therefore if we form an operator that we know will couple two vibrational states, $\psi_i$ and $\psi_j$, then we can determine the symmetry of the operator by determining the symmetry of the vibrational states:

$$
\hat{\mu} \left| \chi_i \right> = \left| \chi_j \right>
$$

If we define the states by their symmetry, this means that:

$$
\hat{\mu} \left| \chi_i \right> = \lambda \left| \chi_i \right>
$$

 - where $\hat{\mu}$ is the dipole moment operator
 - $\left| \chi_i \right>$ is the vibrational state
 - $\lambda$ is the eigenvalue of the vibrational state

The study of the symmetry of transformations on a group is governed by **automorphisms**. An automorphism is a mapping from a group to itself that preserves the group operation. In other words, it is a mapping that preserves the symmetry of the group.

$$
f:G \rightarrow G
$$

The symmetry of an automorphism is a linear map that permutes the basis vectors (which are the group elements) in a way that respects the multiplication rules of the group.

Basically, if the automorphism f takes every element g to itself, then f is called the **identity automorphism** and is denoted by $\hat{E}$.

$$
\hat{E} g = g, \quad   g \in G
$$

An automorphism that permutes the group elements in a way that respects the multiplication rules of the group is called an **inner automorphism**.

In the group $C_2$, if we have an operation that maps the A representation to the B representation, then we can say that the operation is an automorphism:

$$
\hat{\mu} \left| \chi_A \right> = \left| \chi_B \right>
$$

This means that the operation maps E to $C_2$ and maps $C_2$ to E. In terms of the group elements we can represent this automorphism as:

$$
\hat{\mu} \left| E \right> = \left| C_2 \right>
$$


 The automorphism that we are interested in is the automorphism that maps the group to itself. This is called the **identity automorphism** and is denoted by $\hat{E}$. The identity automorphism is the automorphism that maps every element of the group to itself. This means that the identity automorphism is the identity matrix.

$$
\hat{\mu} \left| \chi_i \right> = \left| \chi_i \right>
$$

 - where $\hat{\mu}$ is the dipole moment operator
 - $\left| \chi_i \right>$ is the vibrational state

This is were things get a bit tricky. We are not interested in the symmetry of the state itself but rather how the state transforms under the symmetry operations of the group when transitioning from $\psi$ to $\psi'$. Any state transitioning to itself, $\psi \rightarrow \psi$, will have the same symmetry as the initial state. This means that if we represent the initial state as a ket vector, $\left| \psi \right>$, and the final state as a bra vector, $\left< \psi \right|$, then the symmetry of the transition will be the same as the symmetry of the ket-bra product, $\left| \psi \right> \left< \psi \right|$.
