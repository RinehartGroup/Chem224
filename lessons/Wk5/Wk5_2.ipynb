# Crystal Field Hamiltonian for the d-orbitals
## The orbital angular momentum operator

#### Orbital Angular Momentum in ket notation

The orbital angular momentum operator is given by:

$$
\hat{L} = \hat{r} \times \hat{p} = -i \hbar \hat{r} \times \nabla
$$

Where:
    - $\hat{r}$ is the position operator
    - $\hat{p}$ is the momentum operator
    - $\nabla$ is the gradient operator
    - $\times$ is the vector cross product
    - $i$ is the imaginary unit
    - $\hbar$ is the reduced Planck constant

The orbital angular momentum operator is a vector operator, and so it has three components:

$$
\hat{L}_x = -i \hbar \left( y \frac{\partial}{\partial z} - z \frac{\partial}{\partial y} \right)
$$

$$
\hat{L}_y = -i \hbar \left( z \frac{\partial}{\partial x} - x \frac{\partial}{\partial z} \right)
$$

$$
\hat{L}_z = -i \hbar \left( x \frac{\partial}{\partial y} - y \frac{\partial}{\partial x} \right)
$$

#### Orbital Angular Momentum in matrix notation

The orbital angular momentum operator can be written in matrix notation as:

$$
\hat{L}_x = -i \hbar
\begin{pmatrix}
0 & 0 & 0 \\
0 & 0 & -i \\
0 & i & 0
\end{pmatrix}
$$

$$
\hat{L}_y = -i \hbar
\begin{pmatrix}
0 & 0 & i \\
0 & 0 & 0 \\
-i & 0 & 0
\end{pmatrix}
$$

$$
\hat{L}_z = -i \hbar
\begin{pmatrix}
0 & -i & 0 \\
i & 0 & 0 \\
0 & 0 & 0
\end{pmatrix}
$$

#### Orbital Angular Momentum Commutation Relations

The orbital angular momentum operator has the following commutation relations:

$$
\left[ \hat{L}_i, \hat{L}_j \right] = i \hbar \epsilon_{ijk} \hat{L}_k
$$

Where:
    - $\epsilon_{ijk}$ is the Levi-Civita symbol
      - $\epsilon_{ijk} = +1$ if $ijk$ is an even permutation of $123 \in \{123, 231, 312\}$
      - $\epsilon_{ijk} = -1$ if $ijk$ is an odd permutation of $123 \in \{132, 213, 321\}$


Note that all three components **DO NOT** simultaneously commute with each other!

The Total Angular Momentum operator has the following commutation relations with the three components of the orbital angular momentum operator:

$$
\left[ \hat{L}^2, \hat{L}_i \right] = 0, \quad i = x, y, z
$$

Note that all three components **DO** simultaneously commute with the total angular momentum operator!

#### Orbital Angular Momentum Eigenvalues Example: $l=1$

1. Using bra-ket notation, we can find the total orbital angular momentum quantum numbers using the operator $\hat{L}^2$ because it commutes with all three components of the orbital angular momentum operator:

    $$
    \hat{L}^2 \vert \psi \rangle = l (l+1) \hbar^2 \vert \psi \rangle
    $$


    Where:

    - $\vert \psi \rangle$ is the wavefunction
    - $l$ is the total orbital angular momentum quantum number
    - $\hbar$ is the reduced Planck constant
    - $\hat{L}^2$ is the total orbital angular momentum operator

2. By convention we solve for the z-projection of the orbital angular momentum quantum number $m_l$ using the operator $\hat{L}_z$. For $l=1$, we can find the $z$-component of the orbital angular momentum quantum number using:

    $$
    \hat{L}_z \vert \psi \rangle = m_l \hbar \vert \psi \rangle
    $$

    where:
        
    - $\vert \psi \rangle$ is the wavefunction
    - $m_l$ is the $z$-component of the orbital angular momentum quantum number
    - $\hbar$ is the reduced Planck constant
    - $\hat{L}_z$ is the $z$-component of the orbital angular momentum operator

3. For $l=1$, the possible values of $m_l$ are $-1$, $0$, and $1$.

    This gives us the following possible wavefunctions:

    $$
    \vert \psi \rangle = \vert l=1, m_l=-1 \rangle,
    \quad
    \vert \psi \rangle = \vert l=1, m_l=0 \rangle,
    \quad
    \vert \psi \rangle = \vert l=1, m_l=1 \rangle
    $$

#### Following this logic the possible wavefunctions for $l=2$ are:

$$
\vert \psi \rangle = \vert l=2, m_l=-2 \rangle,
\quad
\vert \psi \rangle = \vert l=2, m_l=-1 \rangle,
\quad
\vert \psi \rangle = \vert l=2, m_l=0 \rangle,
\quad
\vert \psi \rangle = \vert l=2, m_l=1 \rangle,
\quad
\vert \psi \rangle = \vert l=2, m_l=2 \rangle
$$

### Interaction with a Crystal Field of Oh Symmetry

The Hamiltonian for the interaction of a d-orbital with a crystal field of Oh symmetry is given by:

$$
\hat{H} = \hat{H}_0 + \hat{H}_{CF}
$$

and using our kets for l = 2:

$$
\hat{H} =
\begin{pmatrix}
\hat{H}_0 + \hat{H}_{CF} & 0 & 0 & 0 & 0 \\
0 & \hat{H}_0 + \hat{H}_{CF} & 0 & 0 & 0 \\
0 & 0 & \hat{H}_0 + \hat{H}_{CF} & 0 & 0 \\
0 & 0 & 0 & \hat{H}_0 + \hat{H}_{CF} & 0 \\
0 & 0 & 0 & 0 & \hat{H}_0 + \hat{H}_{CF}
\end{pmatrix}
$$

Where:
    - $\hat{H}_0$ is the Hamiltonian for the free ion
    - $\hat{H}_{CF}$ is the Hamiltonian for the crystal field
    - $\hat{H}$ is the total Hamiltonian
    - $\hat{H}_0 + \hat{H}_{CF}$ is the Hamiltonian for the interaction of the d-orbital with the crystal field

So to interact them we can use perturbation theory to find the energy levels of the d-orbital in the crystal field.

### Perturbation Theory

To find the change in energy due to the crystal field we solve the following equation:

$$
\hat{H} \vert \psi \rangle = E \vert \psi \rangle
$$

Where:
    - $\hat{H}$ is the total Hamiltonian
    - $\vert \psi \rangle$ is the wavefunction
    - $E$ is the energy

Which, for the d-orbitals in Oh symmetry, gives us the following matrix block diagonalized matrix:

$$
\begin{pmatrix}
\hat{H}_0 + \hat{H}_{CF} & 0 & 0 & 0 & 0 \\
0 & \hat{H}_0 + \hat{H}_{CF} & 0 & 0 & 0 \\
0 & 0 & \hat{H}_0 + \hat{H}_{CF} & 0 & 0 \\
0 & 0 & 0 & \hat{H}_0 + \hat{H}_{CF} & 0 \\
